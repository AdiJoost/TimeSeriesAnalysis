```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
import warnings
warnings.filterwarnings('ignore')
# Simulate monthly ice cream sales with trend and seasonality
np.random.seed(42)
months = pd.date_range(start='2010-01-01', periods=120, freq='M')
trend = np.linspace(50, 100, 120)
seasonality = 10 + 15 * np.sin(np.linspace(0, 3 * np.pi, 120))
noise = np.random.normal(0, 5, 120)
sales = trend + seasonality + noise

# Create a DataFrame
ice_cream_sales = pd.DataFrame({'Month': months, 'Sales': sales})
ice_cream_sales.set_index('Month', inplace=True)

# Plot the data and forecast
plt.plot(ice_cream_sales['Sales'], label='Historical Sales')
#plt.plot(pd.date_range(start='2020-01-01', periods=12, freq='M'), forecast, label='Forecast', color='red')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

def test_adf(data):
    result = adfuller(data)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])

    if result[1] > 0.05:
        print("The series is non-stationary")
    else:
        print("The series is stationary")

test_adf(ice_cream_sales['Sales'])
```

```python
ice_cream_sales_diff = ice_cream_sales['Sales'].diff().dropna()
test_adf(ice_cream_sales_diff)

ice_cream_sales_diff_diff = ice_cream_sales_diff.diff().dropna()
test_adf(ice_cream_sales_diff_diff)

```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
plot_acf(ice_cream_sales_diff, lags=30, ax=ax[0], title='ACF of Differenced Data')
plot_pacf(ice_cream_sales_diff, lags=30, ax=ax[1], title='PACF of Differenced Data')
plt.show()
```

```python
#ARIMA(2,2,1) 

from statsmodels.tsa.arima.model import ARIMA

# Fit ARIMA(2,2,1) model
model = ARIMA(ice_cream_sales_diff_diff, order=(2, 2, 1))
model_fit = model.fit()

# Print summary
print(model_fit.summary())

```

```python
# Forecast on the twice-differentiated series
forecast_diff_diff = model_fit.forecast(steps=12)

# Undo the second differencing
forecast_diff = pd.Series(0, index=pd.date_range(start=ice_cream_sales.index[-1] + pd.DateOffset(months=1), periods=12, freq='M'))
forecast_diff.iloc[0] = ice_cream_sales_diff.iloc[-1] + forecast_diff_diff[0]
for i in range(1, 12):
    forecast_diff.iloc[i] = forecast_diff.iloc[i-1] + forecast_diff_diff[i]

# Undo the first differencing
forecast = pd.Series(0, index=pd.date_range(start=ice_cream_sales.index[-1] + pd.DateOffset(months=1), periods=12, freq='M'))
forecast.iloc[0] = ice_cream_sales['Sales'].iloc[-1] + forecast_diff.iloc[0]
for i in range(1, 12):
    forecast.iloc[i] = forecast.iloc[i-1] + forecast_diff.iloc[i]

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(ice_cream_sales['Sales'], label='Historical Sales')
plt.plot(forecast, label='Forecasted Sales', color='red')
plt.title('Ice Cream Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
