```python
def plotSomething(df: pd.DataFrame, yAxis: str, xAxis: str = "Date"):
    plt.plot(df[xAxis], df[yAxis])
    ax = plt.gca()
    ax.xaxis.set_major_locator(MaxNLocator(10))
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

Now, we explore the weekly sales

```python
df = pd.read_csv("data/train.csv")
```

```python
print(df.info())
```

```python
storeOne = df[(df["Store"] == 1) & (df["Dept"] == 1)]
print(storeOne.info())
```

```python
plotSomething(storeOne, "Weekly_Sales")
```

```python
storeOne_train = storeOne[:-12]
storeOne_test = storeOne[-12:]
plotSomething(storeOne_train, "Weekly_Sales")
plotSomething(storeOne_test, "Weekly_Sales")
```

```python
#Look vor ARIMA
from statsmodels.tsa.stattools import adfuller

def test_adf(data):
    result = adfuller(data)
    print('ADF Statistic:', result[0])
    print('p-value:', result[1])

    if result[1] > 0.05:
        print("The series is non-stationary")
    else:
        print("The series is stationary")
```

```python
test_adf(storeOne_train["Weekly_Sales"])
```

```python
#differentiating
sales_diff = storeOne_train["Weekly_Sales"].diff().dropna()
test_adf(sales_diff)
#Is stationary -> d = 1
```

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

# Plot ACF and PACF
fig, ax = plt.subplots(2, 1, figsize=(10, 10))
plot_acf(sales_diff, lags=30, ax=ax[0], title='ACF of Differenced Data')
plot_pacf(sales_diff, lags=30, ax=ax[1], title='PACF of Differenced Data')
plt.show()
```

```python
#Arima (2, 1, 2)
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(sales_diff, order=(2,1,2))
model_fit = model.fit()

print(model_fit.summary())
```

```python
forecast_diff = model_fit.forecast(steps=12)

storeOne_train['Date'] = pd.to_datetime(storeOne_train['Date'], format='%Y-%m-%d')
storeOne_test['Date'] = pd.to_datetime(storeOne_test['Date'], format='%Y-%m-%d')
lastDate = storeOne_train["Date"].max()
newDates = pd.date_range(start=lastDate + pd.Timedelta(weeks=1), periods=12, freq="W-FRI")
forecast = pd.DataFrame({"Date": newDates, "Weekly_Sales": [0] * len(newDates)})
print(forecast)
```

```python
PreviousWeekly = storeOne_train["Weekly_Sales"]
print(PreviousWeekly.iloc[-1])
```

```python
storeOne_test
```

```python
forecast_diff
```

```python
forecast.iloc[0, 1] = PreviousWeekly.iloc[-1] + forecast_diff.iloc[0]
for i in range(1, 12):
    forecast.iloc[i, 1] = forecast.iloc[i-1, 1] + forecast_diff.iloc[i]

# Plot the forecast
plt.figure(figsize=(12, 6))
plt.plot(storeOne_train['Date'], storeOne_train['Weekly_Sales'], label='Historical Sales')
plt.plot(forecast["Date"], forecast["Weekly_Sales"], label='Forecasted Sales', color='red')
plt.plot(storeOne_test["Date"], storeOne_test["Weekly_Sales"], label='Actual Sales', color='green')
plt.title('Ice Cream Sales Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
plotSomething(forecast, "Weekly_Sales")
plotSomething(storeOne_test, "Weekly_Sales")
```
